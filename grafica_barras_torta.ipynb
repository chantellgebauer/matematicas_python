Maria tiene 3 manzanas y 6 peras. Mario tiene 5 manzanas y 2 peras. ¿Qué porcentaje de frutas tienen en total?

```python
import pandas as pd
datos = pd.DataFrame({ "manzanas": [3,5], "peras": [6,2]}, index=["Maria", "Mario"])
datos.plot.bar()
```

```python
datos["frutas en total"]= datos["manzanas"] + datos["peras"]
datos.plot.pie(y="frutas en total", autopct='%1.1f%%')
```
